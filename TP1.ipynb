```python
from PIL import Image, ImageDraw

def create_italy_flag(width, height):
    img = Image.new('RGB', (width, height), 'white')
    draw = ImageDraw.Draw(img)
    stripe_width = width // 3
    colors = [(0, 146, 70), (255, 255, 255), (206, 17, 38)]
    for i, color in enumerate(colors):
        left = i * stripe_width
        right = (i + 1) * stripe_width
        draw.rectangle([left, 0, right, height], fill=color)
    draw.rectangle([0, 0, width - 1, height - 1], outline='black')
    return img

# Crez les images et enregistrez-les
italy_flag = create_italy_flag(300, 200)
italy_flag.save('/content/italy_flag.png')
italy_flag
```
